## Description: 

This Python program is a simple number guessing game built using a while loop. The computer randomly selects a secret number, and the player keeps guessing until they find the correct one. After each guess, the program gives hints such as “guess high” or “guess low,” helping the player reach the right answer. The use of a while loop ensures that the game continues running until the correct guess is made, making it an interactive and fun way to practice loops, conditionals, and user input in Python.

## Why using while loop ?

Since we don't know how many attempts the user will take, so we use a while loop.

```python
import random
```

```python
jackpot = random.randint(1,100)

guess = int(input("Guess the number:"))
counter = 0

while guess != jackpot:
    if guess < jackpot:
        print("guess higher")
    else:
        print("guess lower")
        
    guess = int(input("Guess the number:"))
    counter += 1 
    
print("correct guess")
print(f"you took {counter} attempts")

```
